```python
import numpy as np
import librosa as lb
import matplotlib.pyplot as plt
import IPython.display as ipd
import scipy.io as sio
import time
```

```python
def loadAudioCalcSTFT(audiofile):
    x, sr = lb.core.load(audiofile)
    S = lb.core.stft(x)
    Smag = np.abs(S) # magnitude
    Sphase = np.angle(S) #phase
    N = 1025 #librosa default window size
    return x, sr, N, S, Smag, Sphase
```

```python
x, sr, N, S, Smag, Sphase = loadAudioCalcSTFT('data/aapte_rec3.wav')

plt.figure(1, figsize=(4,160))
plt.imshow(Smag[:,0:500], origin = 'lower', cmap = 'jet')
```
